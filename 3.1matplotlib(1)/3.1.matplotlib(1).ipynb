## 3. 绘制一个散点图

步骤：生成画布，生成子图->在子图上画图

```python
fig_scatter, ax_scatter = plt.subplots()
#定义了一个画布，变量名为fig_scatter，在这个画布上定义了一个子图，变量名为ax_scatter.子图的个数为1行1列共1个,改括号里的值即可修改子图个数
```

```python
ax_scatter.scatter(x, y)
```

```python
fig_scatter #调用画布，可以展示画布上绘制的图形
```

```python
fig_scatter.savefig('a.jpg') #图像的保存。保存的画布Figure，不是其他的
```

保存图形的类型，是直接由文件名决定的。

想保存成保留透明度的图，可以保存成png

## 案例分析-中国各省GDP、电力消费、人口之间的关系

```python
data=np.loadtxt("data.csv",dtype="int",delimiter=",") 
print(data)
#数据来源，国家统计局：https://data.stats.gov.cn/easyquery.htm?cn=A01
#第一行是各省2018年的GDP（亿元），第二行是2018年的用电量（亿千瓦时），第三行是人口（万人）
```

```python
fig_data,ax_data=plt.subplots()
```

```python
ax_data.scatter(data[0], data[1])
```

```python
fig_data
```

### 修改点的形状、颜色、大小

```python
ax_data.cla() #清空子图上的内容
fig_data
```

```python
ax_data.cla()
ax_data.scatter(data[0], data[1],marker='*',c='green',s=200)
fig_data
```

marker的形状列表：https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers

### 进阶：用大小和颜色的改变表示一个数据维度

```python
ax_data.cla() #把前边画的图清除掉
pic=ax_data.scatter(data[0], data[1],marker='o',c='#00ab23',s=data[2]/20) 
#pic这个变量是一个图形，留待后用
fig_data
```

### 添加图标题和轴标题

```python
ax_data.set_title('GDP-Electricity consumption-Population')
```

```python
fig_data
```

```python
ax_data.set_xlabel('GDP (100 Million RMB)')
ax_data.set_ylabel('Electricity consumption (100 Million kWh)')
```

```python
fig_data
```

### 调整坐标刻度

```python
ax_data.set_xlim(-10000,110000) #调整的是横轴的左右极限
ax_data.set_ylim(-100,8000) #调整的是纵轴的左右极限
fig_data
```

```python
ax_data.set_xticks([0,25000,50000,60300,75000,100000]) #刻度设置接收的是数组/list
fig_data
```

```python
ax_data.set_yticks(np.arange(0,8001,2000))
fig_data
```

### 添加网格线

```python
ax_data.grid(True) #接收布尔值，表示开关
fig_data
```

### 添加图例

```python
handles, labels = pic.legend_elements(prop="sizes") 
#以size属性制作图例，“size”可以换成"color"等
ax_data.legend(handles, labels, borderaxespad=0., title="Populations(0.2 million)",bbox_to_anchor=(1.05, 1), loc='upper left') 
#因为前边画人数时除以了50
fig_data
```

```python
# 注意保存图片格式的选择，jpg是不透明的，png是透明的，这两者都是位图，pdf是矢量图，可以任意放大缩小不改变形状
fig_data.savefig('20210310.pdf', bbox_inches = 'tight')
# bbox_inches = 'tight'这一句用来把图例所在的bbox也保存进去
```

数据分析结论：

1. GDP越高，用电量也越高
2. 人口越少，GDP越低
3. 人口越少，用电量越少
