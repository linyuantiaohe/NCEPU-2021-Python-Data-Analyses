## 2. 折线图

折线图（line chart）或曲线图（curve chart）是由许多的资料点用直线连接形成的统计图表，若看多个资料点之间的连线，会是折线。折线图是许多领域都会用到的基础图表，折线图类似散布图，不过折线图以X轴为基础，将X轴上相邻的资料点之间用直线连接。折线图常用来观察资料在一段时间之内的变化（时间序列），因此其X轴为时间，这种折线图又称为趋势图。

定义为：Axes.plot(self, *args, scalex=True, scaley=True, data=None, **kwargs)

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html

数据来源：国家统计局https://data.stats.gov.cn/easyquery.htm?cn=A01

```python
# 2017-2019的月度全国火力发电量数据（亿千瓦时），1月2月不统计，每年有10个数据
power_genneration_monthly=np.array([3960.8,3521.8,3552.7,3709.7,4334.3,4310,3622.7,3475,3762.3,4416.9,4017.4,3784.7,3921.5,3940.4,4534.9,4601.7,3787.3,3629.4,4050.7,4775.9,4159.7,3886.2,3830.6,4051.9,4562.5,4728.5,4149.6,3969.5,4325.1,5081.7])
print(power_genneration_monthly)
```

```python
fig_line,ax_line=plt.subplots()
ax_line.plot(power_genneration_monthly)
```

用for循环改写上述写法

### 2.2 组合点+线

#### 2.2.1 方法一：在画布上依次画线和点

```python
fig_linedot,ax_linedot=plt.subplots()
for  i,y in zip(power_generation_year_month,year):
    ax_linedot.plot(month,i,label='%d'%y)
    ax_linedot.scatter(month,i,label='%d'%y)  #直接在画布上用scatter命令画出点
ax_linedot.legend()
```

#### 2.2.2 方法二：点线图

```python
ax_linedot.cla()
colorlist=['black','blue','red']
for  i,y,c in zip(power_generation_year_month,year,colorlist):
    ax_linedot.plot(month,i,'-.^',color=c,label='%d'%y) #紧跟数据的位置，可以定义图形的形状
ax_linedot.legend()
fig_linedot
```

#### 2.2.3 图形的完善

```python
ax_linedot.set_xlabel('月份')
ax_linedot.set_xticks(np.arange(3,13,1))
ax_linedot.set_ylabel('火电发电量（亿度）')
ax_linedot.set_title('2017-2019年月度火力发电数据')
ax_linedot.grid(True) #打开网格
fig_linedot
```

```python
fig_linedot.savefig('2017-2019年月度火力发电数据.pdf')
```

## 3. 柱状图

又叫条形统计图、直方图等。在统计学中，直方图（英语：Histogram）是一种对数据分布情况的图形表示，是一种二维统计图表，它的两个坐标分别是统计样本和该样本对应的某个属性的度量，以长条图（bar）的形式具体表现。因为直方图的长度及宽度很适合用来表现数量上的变化，所以较容易解读差异小的数值。

定义为：Axes.bar(self, x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html

### 3.1 基本柱状图

```python
fig_bar,ax_bar=plt.subplots()
ax_bar.bar(month3,power_genneration_monthly)
ax_bar.set_xticklabels(month3, rotation=90)
```

### 3.2 多维柱状图

例如：3年月度数据对比，需要使用并列柱状图

提示：比较复杂，记住思路，用的时候再仔细研究即可

```python
total_years=3  #即我们一共需要3个柱子
total_width=0.8  #所有柱子加起来总共的相对宽度
single_width = total_width / total_years  #每个柱子单独的相对宽度

month=np.arange(3,13,1)  #横坐标，即left值
print(month)
month=month-(total_width - single_width)/(total_years-1)  #奇数个柱除的时候这样，偶数个柱除的时候就total_width/total_years
print(month)
```

```python
ax_bar.cla()
ax_bar.bar(month,power_generation_year_month[0],width=single_width, label='2017')
ax_bar.bar(month+single_width,power_generation_year_month[1],width=single_width, label='2018')
ax_bar.bar(month+2*single_width,power_generation_year_month[2],width=single_width, label='2019')
ax_bar.legend()
fig_bar
```

用for循环改写上述写法

```python
ax_bar.cla()
for i in range(3):
    ax_bar.bar(month+i*single_width,power_generation_year_month[i],width=single_width, label='%d'%(2017+i))
ax_bar.legend()
fig_bar
```

### 3.3 完善图形

```python
ax_bar.set_xlabel('月份')
ax_bar.set_xticks(np.arange(3,13,1))
ax_bar.set_ylabel('火电发电量（亿度）')
ax_bar.set_title('2017-2019年月度火力发电数据')
fig_bar
```

```python
fig_bar.savefig('柱状图.pdf')
```

### 4.1 分析同比（year-on-year ）增长量

同比增长是第N+1年的t月相比第N年的t月的增长，其内生考虑了一年内的周期性。

```python
annual_yoy_increase=np.zeros((2,10))
annual_yoy_increase[0]=power_generation_year_month[1]-power_generation_year_month[0]  #2018年对于2017年的同比增长量
annual_yoy_increase[1]=power_generation_year_month[2]-power_generation_year_month[1]  #2019年对于2018年的同比增长量
print(annual_yoy_increase)
```

```python
total_years=2  #即我们一共需要2个柱子
total_width=0.8  #所有柱子加起来总共的相对宽度
single_width = total_width / total_years  #每个柱子单独的相对宽度

month=np.arange(3,13,1)  #横坐标，即left值
month=month-total_width /total_years  #奇数个柱除的时候这样，偶数个柱除的时候就total_width/total_years
```

```python
fig_yoy_increase,ax_yoy_increase=plt.subplots()
for i in range(2):
    ax_yoy_increase.bar(month+i*single_width,annual_yoy_increase[i],width=single_width, label='%d'%(2018+i))
ax_yoy_increase.legend()
```

```python
ax_yoy_increase.set_xlabel('月份')
ax_yoy_increase.set_ylabel('平均同比增长量（亿度）')
ax_yoy_increase.set_title('2017-2019年月度火力发电同比增长量')
fig_yoy_increase
```

### 4.2 分析同比增长率

求个两年各月份平均同比增长率

增长率=(x2-x1)/x1

```python
average_yoy_increase_rate=((power_generation_year_month[1]-power_generation_year_month[0])/power_generation_year_month[0]+(power_generation_year_month[2]-power_generation_year_month[1])/power_generation_year_month[1])/2
```

```python
print(average_yoy_increase_rate)
```

```python
fig_yoy_increase_rate,ax_yoy_increase_rate=plt.subplots()
```

```python
month=np.arange(3,13,1)
```

```python
ax_yoy_increase_rate.cla()
ax_yoy_increase_rate.bar(month,average_yoy_increase_rate)
ax_yoy_increase_rate.set_xticks(np.arange(3,13,1))
ax_yoy_increase_rate.set_xlabel('月份')
ax_yoy_increase_rate.set_ylabel('平均同比增长率')
ax_yoy_increase_rate.set_title('2017-2019年月度火力发电同比增长率')
fig_yoy_increase_rate
```

```python
fig_yoy_increase_rate.savefig('平均同比增长率.pdf')
```

### 4.3 分析环比（month-on-month）增长率

环比即N+1月相比N月的增长率。不考虑年内的周期性。

思考：如何计算环比增长率？

```python
print(power_genneration_monthly)
```

通过for循环来实现

```python
mom_increase_rate=np.zeros(30)
print(mom_increase_rate)
```

```python
range(power_genneration_monthly.size-1)
```

```python
for i in range(power_genneration_monthly.size-1):
    mom_increase_rate[i+1]=(power_genneration_monthly[i+1]-power_genneration_monthly[i])/power_genneration_monthly[i]
```

```python
print(mom_increase_rate)
```

```python
mom_increase_rate=mom_increase_rate.reshape(3,10)
```

```python
mom_increase_rate
```

```python
#数组.mean()，无参数，返回所有元素的均值，加上axis参数，可以返回行均值或者列均值
average_mom_increase_rate=mom_increase_rate.mean(axis=0)
```

```python
average_mom_increase_rate
```

```python
fig_mom_increase_rate,ax_mom_increase_rate=plt.subplots()
ax_mom_increase_rate.cla()
ax_mom_increase_rate.bar(month,average_mom_increase_rate)
ax_mom_increase_rate.set_xticks(np.arange(3,13,1))
ax_mom_increase_rate.set_xlabel('月份')
ax_mom_increase_rate.set_ylabel('平均环比增长率')
ax_mom_increase_rate.set_title('2017-2019年月度火力发电环比增长率')
```

环比并不适宜于分析周期性较强的数据

## 基于twinx()函数的双坐标轴

```python
x, y = np.random.rand(2,30)
x.sort()
y.sort()
fig, ax_1 = plt.subplots()
ax_1.plot(x, y,label='line1')
ax_1.plot(5*y,3*x,label='line2')
ax_1.legend()
```

```python
#新建一个子图，但和原来的子图ax_1公用一个x轴
ax_2 = ax_1.twinx()
print(type(ax_2))
print(type(ax_1)==type(ax_2))
fig
```

```python
ax_1.cla()
ax_2.cla()
ax_1.plot(x, y,label='line1',c='b')
ax_2.plot(5*y,3*x,label='line2',c='r')
ax_1.set_ylabel('ax_1的y轴')
ax_2.set_ylabel('ax_2的y轴')
fig
```

```python
ax_1.legend()
ax_2.legend()
fig
```

```python
fig.legend()
fig
```

```python
'''
可以通过remove（）去除图例
要注意legend后边有个下划线_
下划线后边接的.remove()
'''
ax_1.legend_.remove()
ax_2.legend_.remove()
fig
'''
另外要注意，remove只能在有图例的情况下使用，已经去除的图例不能被2次去除
再次运行就会报错
'''
```

另外ncol参数是定义图例的列数的，默认为1.
