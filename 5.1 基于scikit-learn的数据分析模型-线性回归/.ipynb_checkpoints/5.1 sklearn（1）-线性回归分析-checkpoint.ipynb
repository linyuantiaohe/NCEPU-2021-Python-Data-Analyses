## 最小二乘法线性回归

**线性回归**`sklearn.linear_model.LinearRegression`

目标值 y 是输入变量 x 的线性组合。 数学概念表示为：如果$\hat{y}$
 是预测值，那么有：

$\hat{y}(w, x) = w_0 + w_1 x_1 + ... + w_p x_p$

在整个模块中，我们定义向量 $w = (w_1,..., w_p)$ 作为 $coef_$ ，定义 $w_0$ 作为 $intercept_$ 。

**最小二乘法（Ordinary Least Squares，OLS）**

线性回归拟合一个带有系数 $w = (w_1, ..., w_p)$ 的线性模型，使得数据集实际观测数据和预测数据（估计值）之间的残差平方和最小。其数学表达式为:

$$\min_{w} || X w - y||_2^2$$

**scikit-learn实现**

LinearRegression 会调用 $fit$ 方法来拟合数组 $X， y$ 并且将线性模型的系数 $w$ 存储在其成员变量 $coef_$ 中:
![lm.fit](lm.fit.png)

### 例子1：单变量线性回归

```python
from sklearn import linear_model as lm
```

由于scikit-leaen这个包中的功能很多，我们可以只调用我们分析时要使用的那部分功能，可以减少代码量

```python
reg = lm.LinearRegression()  #构建一个线性回归对象
type(reg)
```

构建一组用于回归的数据

```python
x = np.arange(16).reshape(-1,1)
```

```python
x
```

```python
y = x*3+3
y
```

```python
reg.fit(x,y)
```

```python
print('系数为：',reg.coef_,'\n截距为：',reg.intercept_)
```

### 例子2：北京市2000-2015年GDP和人口的关系

```python
beijing=pd.read_excel('北京GDP和人口.xlsx',index_col=0)
beijing
```

```python
fig,ax=plt.subplots()
ax.scatter(beijing['POP'],beijing['GDP'])
```

看图发现随着人口增加，GDP是增长的。

我们**假设**GDP是人口的线性函数，即
$$H: GDP=a\times POP+b$$
我们使用上述的最小二乘法线性回归

```python
reg_pop_to_gdp = lm.LinearRegression() 
```

将数据转换成fit接收的数据

```python
beijing['POP']
```

```python
beijing['POP'].values
```

```python
beijing['POP'].values.reshape(-1,1)
```

```python
x=beijing['POP'].values.reshape(-1,1)
y=beijing['GDP'].values.reshape(-1,1)
x
```

```python
reg_pop_to_gdp.fit(x,y)
```

```python
print('系数为：',reg_pop_to_gdp.coef_,'\n截距为：',reg_pop_to_gdp.intercept_)
```

把拟合结果画在图上

```python
y_hat=x*reg_pop_to_gdp.coef_+reg_pop_to_gdp.intercept_
y_hat
```

```python
ax.plot(x,y_hat,color='green')
fig
```

那么如果人口达到了3000万或者3500万，GDP会是多少？

可以使用predit功能实现

```python
reg_pop_to_gdp.predict([[3000],[3500]]) #注意在输入自变量时，要以二维数组的形式，即两层中括号嵌套
```

即北京人口3000万，那么GDP会达到38932.85344095亿元；人口达到3500万，GDP会达到49965.87763386亿元

真实数据：

|     | 2019    | 2018    | 2017    | 2016    |
|-----|---------|---------|---------|---------|
| 人口  | 2154    | 2154    | 2171    | 2173    |
| GDP | 35445.1 | 33106.0 | 29883.0 | 27041.2 |

### 例子3：多元线性回归

```python
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
# y = 1 * x_0 + 2 * x_1 + 3
y = np.dot(X, np.array([1, 2])) + 3
print(y)
```

```python
reg = lm.LinearRegression().fit(X, y)
print("系数：",reg.coef_,"\n截距：",reg.intercept_)
```

### 例子4：华北地区其他省份GDP对北京GDP的影响

```python
northchina_gdp=pd.read_excel('华北地区各省GDP.xlsx',index_col=0)
northchina_gdp
```

我们提出**假设**北京的GDP受其他华北地区省份GDP的影响
$$H: GDP_{bj}=w_1\times GDP_{tj}+w_2\times GDP_{hb}+w_3\times GDP_{sx}+w_4\times GDP_{nm}+b$$
我们使用上述的最小二乘法线性回归

首先把数据转成scikit-learn能接受的类型，即自变量们构成一个N行X列的二维数组，每一列是一个变量，每一行是一组观测样本。因变量构成一个N行1列的数组。

```python
X=northchina_gdp[['天津市', '河北省', '山西省', '内蒙古自治区']].values
y=northchina_gdp[['北京市']].values
```

```python
X
```

```python
y
```

可以先画图观察每个自变量与因变量之间的相关性

```python
northchina_gdp.columns.values[1]
```

```python
fig_nc,ax_nc=plt.subplots(1,4,figsize=(16,4))
for i in range(4):
    ax_nc[i].plot(X.T[i],y.T[0])
    ax_nc[i].set_title(northchina_gdp.columns.values[i+1]+'-北京')
```

```python
reg_northchina = lm.LinearRegression().fit(X, y)
```

```python
print("系数：",reg_northchina.coef_,"\n截距：",reg_northchina.intercept_)
```

```python
y_hat=np.dot(X,reg_northchina.coef_.reshape(4,1))+reg_northchina.intercept_
```

```python
year=np.arange(2018,1992,-1)
print(year)
fig1,ax1=plt.subplots()
ax1.plot(year,y,c='red')
ax1.plot(year,y_hat,c='green')
```

## 机器学习模型评价

通过机器学习做出回归模型后，一个重要的用途是用来做预测。那么模型的预测能力就需要做出评估。

比如我们想要评估例子4中模型的预测能力，可以把1993-2018年的数据分为2部分，一部分用来学习，一部分用来检验学习成果。

以下我们取1993-2012，20年的数据用来学习，2013-2018 6年的数据用来检验。

```python
X_study=X[6:]
y_study=y[6:]
X_test=X[0:6]
y_test=y[0:6]
```

```python
reg_study=lm.LinearRegression().fit(X_study, y_study)
```

```python
y_predict=reg_study.predict(X_test)
```

```python
fig3,ax3=plt.subplots()
ax3.plot(y_test,c='r')
ax3.plot(y_predict,c='b')
```

可以看到，基于1993-2012这20年的数据学习来的模型，预测的结果偏低。

**sci-kit learn这个包提供了一个score功能，用来给模型打分**

返回的是$R^2 =(1 - u/v)$, 其中 `u=((y_true - y_pred) ** 2).sum()` ,`v=((y_true - y_true.mean()) ** 2).sum()`
取值区间为$(-\infty,1.0]$ 

![score definition](score.png)

```python
reg_study.score(X_test,y_test)
```

```python
reg_study.score(X_study,y_study)
```

可见这个模型预测的准确性很差，说明（1）这个模型不合适（2）在2012年之后GDP变化规律发生了重大变化

过度拟合：学习数据拟合程度很高，预测数据拟合程度很低

## 实例

```python
data=pd.read_excel('全国GDP和居民消费水平.xlsx',index_col=0)
data
```

```python
X_study=data['GDP'].values[4:].reshape(-1, 1)
y_study=data['居民消费水平'].values[4:].reshape(-1, 1)
X_test=data['GDP'].values[0:4].reshape(-1, 1)
y_test=data['居民消费水平'].values[0:4].reshape(-1, 1)
```

```python
reg_ti_gdp=lm.LinearRegression().fit(X_study, y_study)
y_predict=reg_ti_gdp.predict(X_test)
```

```python
reg_ti_gdp.score(X_test,y_test)
```

预测效果不错

```python
fig3,ax3=plt.subplots()
ax3.plot(data['GDP'],data['居民消费水平'],'r-o')
ax3.plot(data['GDP'],reg_ti_gdp.predict(data['GDP'].values.reshape(-1, 1)),'b--o')
ax3.set_xlabel('GDP(亿元)')
ax3.set_ylabel('居民消费水平(元)')
```
