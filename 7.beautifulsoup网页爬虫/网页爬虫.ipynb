## 爬虫概念

### 什么是爬虫？

爬虫就是通过编写程序模拟浏览器上网，然后让其去互联网上抓取数据的过程。

### 为什么要爬虫？

1. 网络时代，数据的形式越来越多，数字、文字、图形、音频、视频等等具有分析价值，且许多数据都保存在互联网上。
2. 要进行大数据分析，首先要有数据源，而学习爬虫，可以让我们轻松采集这些分散在互联网上的数据。

**python实现和处理爬虫语法简单，代码优美，支持的模块繁多，学习成本低，具有非常强大的框架。**

### 爬虫的分类

#### 通用爬虫

通用爬虫是搜索引擎（Baidu、Google、Yahoo等）“抓取系统”的重要组成部分。主要目的是将互联网上的网页下载到本地，形成一个互联网内容的镜像备份。  简单来讲就是尽可能的；把互联网上的所有的网页下载下来，放到本地服务器里形成备分，在对这些网页做相关处理(提取关键字、去掉广告)，最后提供一个用户检索接口。 

#### 聚焦爬虫

聚焦爬虫是根据指定的需求抓取网络上指定的数据。例如：获取豆瓣上电影的名称和影评，而不是获取整张页面中所有的数据值。

### 爬虫的流程

找到网页地址（URL）->下载网页内容->提取其中需要的部分->保存数据。

### 实例：北京天气查询

举例：http://www.weather.com.cn/weather/101010100.shtml  查询北京7日天气
网页上有很多广告，我们可以通过爬虫以只获得我们需要的信息。

## 爬虫的实现

### 网页结构

一般的网页是用html语言编写的，再通过浏览器解释为我们看到的网页。

html语言是一种标签语言，和markdown一个原理，只不过html的标签更丰富。

`<a></a>`每个标签都有一个起点和一个终点，终点和起点标签名相同，多了个斜杠

HTML入门：https://www.w3school.com.cn/html/html_jianjie.asp

### 爬虫第一步：下载网页

python实现方法：urllib.request包(python自带）

官网：https://docs.python.org/3/library/urllib.request.html

urllib.request.urlopen(url, data=None, [timeout, ]*, cafile=None, capath=None, cadefault=False, context=None)

```python
from urllib.request import urlopen  #urlopen模块顾名思义可以打开网址
```

```python
a=urlopen("http://www.weather.com.cn/weather/101010100.shtml")
a
```

### 爬虫第二步：解析网页

我们得到是网页，需要的是网页上的内容，而不需要那些标签。因此需要提取没有标签的内容。

但网页内容通过标签排列，可以通过标签对内容进行定位。

python实现方法: beautifulsoup包

官方文档：https://www.crummy.com/software/BeautifulSoup/bs4/doc/

中文文档：https://www.crummy.com/software/BeautifulSoup/bs4/doc/index.zh.html  （较英文文档落后若干版本）

安装方法：
`pip install beautifulsoup4`

```python
from bs4 import BeautifulSoup #从bs4包中导入Beautifulsoup模块
```

```python
b=BeautifulSoup(a)
```

```python
b
```

即通过BeautifulSoup解析urlopen打开的网址，可以得到网页全部代码

```python
type(b) #是一个BeautifulSoup类型的对象
```

得到网页代码后，需要从中筛选出我们想要的标签。标签的确定需要手动操作。推荐：
1. 使用chrome/Firefox等浏览器查询网页源码，确定大概范围。
2. 通过VScode等文本编辑器将标签按层级添加缩进，以确定精确位置。

BeautifulSoup对象可以通过find命令搜索所需标签，得到标签包围的内容。

使用方法为bs4.BeautifulSoup.find(标签名，{标签属性：{标签属性值1，标签属性值2，...}})

```python
c=b.find("ul",{'class':{'t clearfix'}})
c
```

```python
type(c)
```

BeautifulSoup可以嵌套使用find，即得到的Tag对象还可以继续执行find

**接下来，我们找到有用的信息**

分析c这段网页代码可以看出，每一天的天气信息都是按同一格式排列的，包含在一个`<li>`标签中，一共有7天，即有7个标签

可以使用findAll命令来找到所有这个名字的标签，保存成一个list

```python
d=c.findAll("li")
d
```

```python
type(d)
```

```python
d[0]
```

Beautifulsoup的Tag对象支持get_text()来去掉标签，仅获取正文部分

```python
e=d[0].get_text()
e
```

对字符串执行replace命令，可以将字符串中的\n删除

'str'.replace('待替换内容'，'替换内容')

```python
e.replace('\n',' ')  #将换行符换成空格
```

```python
#综上，我们输出结果：
print('北京天气:\n')
for i in d:
    print(i.get_text().replace('\n',' '))
```

接下来，我们一次输出region这个list里所有的地址的地名和天气。

首先构建网址：

```python
i=region[0]
url="http://www.weather.com.cn/weather/"+str(i)+".shtml"
url
```

用for循环来构建地址并输出：

```python
from urllib.request import urlopen 
from bs4 import BeautifulSoup 
region=[101010100,101130101,101010300,101011200]
for i in region:
    url="http://www.weather.com.cn/weather/"+str(i)+".shtml"
    a=urlopen(url)
    b=BeautifulSoup(a)
    c1=b.find("div",{'class':{"crumbs fl"}})
    regionname=c1.get_text().replace('\n','').replace('>',' ')
    c2=b.find("ul",{'class':{'t clearfix'}}).findAll("li")
    print(c1.get_text().replace('\n','').replace('>',' '),'天气:\n')
    for j in c2:
        print(j.get_text().replace('\n',' '))
    print('-----------------------------')
```
