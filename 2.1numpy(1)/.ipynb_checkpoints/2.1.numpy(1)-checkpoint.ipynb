```python
if a<b:
    print(a)
    print(b)
if a<b:
 print(a)
 print(b)
```

### 2.5 通过索引访问数组

一维数组的访问，先生成一个长度为10的数组

```python
arr = np.arange(10)
print(arr)
```

```python
arr[3] #数组的第4个
```

```python
arr[3:6] #数组中编号为3，4，5的元素，不包括编号为6的，即冒号这样的切片是左闭右开的
```

```python
arr[:3] #冒号左边省略的话即从编号为0的元素开始
#arr[0:3] 等效
```

```python
arr[6:] #冒号右边省略的话则到最后一个元素为止
#和arr[6:9]不等价，arr[6：正无穷]
```

```python
arr[6:9]
```

```python
arr[-1]  #负数索引即倒数第几个元素，-1即为倒数第1个
```

```python
arr[-5:-2] #负数也可以做切片，同样左闭右开
```

```python
arr[0:7:2] #两个冒号代表的是从0到7以2为步长取值，即隔一个取一个
```

```python
arr[7:0:-2] #以负数为步长的话可以倒着取值
```

```python
arr[-2:-5:1]  #在切片时，应注意，步长默认为1，左断点应比右端点小；只有当步长为负数时，左端点应比右端点大
```

多维数组的访问，先生成一个3X5二维数组

```python
arr = np.array([[1, 2, 3, 4, 5],[4, 5, 6, 7, 8], [7, 8, 9, 10, 11]])
print('创建的二维数组为：',arr)
```

```python
arr[0][1]
```

```python
print('索引结果为：',arr[0,3:5])  #索引第1行中第3到4列的元素，[]中的逗号用来区分第一个维度和第二个维度，三维数组的话就有2个逗号分隔
```

```python
print('索引结果为：',arr[1:,2:]) #索引第2行到最后一行中第3列到最后一列的元素
```

```python
print('索引结果为：',arr[:,2])  #索引第2列的元素，逗号前的冒号表示所有的行
```

多维数组的索引（使用整数和布尔值索引访问数据）

```python
print('索引结果为：',arr[((0,1),(1,2))]) #从两个序列的对应位置取出两个整数来组成下标：arr[0,1], arr[1,2]
```

```python
print('索引结果为：',arr[1:,(0,2,3)])  #索引第2行到最后一行中第0、2、3列的元素
```

```python
mask = np.array([1,0,1],dtype = np.bool) #mask是一个布尔数组，它索引第1、3行中第2列的元素
print(mask)
```

```python
print('索引结果为：',arr[mask,2])
```
